# **Dataset Loading:**
- The Fashion MNIST dataset is loaded using TensorFlow's Keras API, which provides training data, training labels, test data, and test labels.

```python
from tensorflow.keras.datasets import fashion_mnist

(x_train,y_train), (x_test, y_test) = fashion_mnist.load_data() 

classes = ['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
print("Download Completed")
```

### Check GPU

```python
from tensorflow.python.client import device_lib

print(device_lib.list_local_devices())
```

### Inspect data

```python
print(f"Number of samples in X_train:{len(x_train)}")
print(f"Number of samples in X_test:{len(x_test)}")
print(f"Number of samples in Y_train:{len(y_train)}")
print(f"Number of samples in Y_test:{len(y_test)}")

print(f"Shape of training images:{x_train[0].shape}")
print(f"Shape of testing images:{x_test[0].shape}")

print(f"Shape of training labels:{y_train.shape}")
print(f"Shape of testing labels:{y_test.shape}")
```

### Visualizing 50 samples

```python
import matplotlib.pyplot as plt

figure = plt.figure()
plt.figure(figsize=(16,10))

num_of_images = 50 

for index in range(1, num_of_images + 1):
    class_names = classes[y_train[index]]
    plt.subplot(5, 10, index).set_title(f'{class_names}')
    plt.axis('off')
    plt.imshow(x_train[index], cmap='gray_r')
```

# **Data Preprocessing:**
- The images in the dataset have a resolution of 28x28 pixels, which are flattened to create an input shape for the CNN.
- One-hot encoding is used for the target labels, converting them into binary vectors for each class.


```python
img_rows = x_train[0].shape[0]
img_cols = x_train[0].shape[1]

input_shape = (img_rows, img_cols, 1)
```

### One hot encoding for the dataset

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
num_classes = y_train.shape[1]
print(f"Number of classes available: {num_classes}")
```

# **CNN Model Building:**
- A Sequential model is used to build the CNN architecture.
- The model architecture is not detailed in this study note, but it is expected to be similar to the CNN used for the MNIST dataset.
- The model is compiled with the stochastic gradient descent (SGD) optimizer and the categorical cross-entropy loss function.


### Import Dependencies

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, MaxPooling2D, Conv2D, Dropout
from tensorflow.keras.optimizers import SGD
from tensorflow.keras import backend as K
```

```python
model = Sequential()

model.add(Conv2D(32, kernel_size=(5,5), activation='relu', input_shape = input_shape))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(128, kernel_size=(5,5), activation='relu'))
model.add(Flatten())
model.add(Dense(128,  activation='relu', ))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer= SGD(0.001, momentum=0.9),
             metrics = ['Accuracy'])

print(model.summary())
```

**Model Training:**
- The compiled CNN model is trained on the Fashion MNIST dataset for 15 epochs.
- Training accuracy and validation accuracy are monitored to analyze model performance.


```python
batch_size = 32
epochs = 25

history = model.fit(x_train,
                    y_train,
                    batch_size = batch_size,
                    epochs = epochs,
                    verbose = 1,
                    validation_data = (x_test, y_test))

score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

# **Overfitting:**
- An important observation during training is the possibility of overfitting.
- Overfitting occurs when the model performs well on the **training data** but fails to generalize to **unseen data**.
- Overfitting is evident as the training accuracy increases while the validation accuracy plateaus and may even slightly decrease.


**Regularization Techniques:**
- To mitigate overfitting, regularization techniques are introduced.
- Early stopping is one potential technique, where training stops if the validation loss or accuracy doesn't improve for a certain number of epochs.
- Other regularization methods like L1 and L2 regularization, dropout, and batch normalization could also be used to improve generalization.

**Regularization Techniques:**
1. **L2 Regularization:**
   - Also known as weight decay.
   - Introduces a penalty term to the loss function that discourages large weights in the model.
   - Helps prevent overfitting by limiting the complexity of the model.

2. **Data Augmentation:**
   - Involves applying random transformations to the training images, such as rotations, shifts, and zooms.
   - Increases the diversity of the training data, allowing the model to generalize better to unseen data.

3. **Batch Normalization:**
   - Normalizes the inputs of each layer in the model during training.
   - Helps stabilize and accelerate training, reducing the need for careful tuning of learning rates.

4. **Dropout:**
   - Randomly drops out (sets to zero) a fraction of neurons during training.
   - Prevents co-adaptation of neurons and helps the model become more robust.



# Preprocess using ImageDataGenerator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
from tensorflow.keras import backend as K

x_train = x_train.reshape(x_train.shape[0],28,28,1)
x_test = x_test.reshape(x_test.shape[0],28,28,1)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

img_rows = x_train[0].shape[0]
img_cols = x_train[0].shape[1]

input_shape = (img_rows, img_cols, 1)

x_test /= 255.0
```

### OneHotEncoding

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

num_classes = y_train.shape[1]
print(f"Number of classes available: {num_classes}")
```

**Model Implementation:**
- The CNN model is implemented using Keras Sequential API with regularized layers and batch normalization added.
- Image Data Generator from Keras is used to perform data augmentation.
- The model is compiled with a specified optimizer, loss function, and metrics.



### Import dependencies including apis for regularizasion

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, MaxPooling2D, Conv2D, Dropout, BatchNormalization
from tensorflow.keras.optimizers import SGD
from tensorflow.keras import backend as K
from tensorflow.keras import regularizers
```

### Create the model with the regularizers defined

```python
model = Sequential()



model.add(Conv2D(32, kernel_size=(5,5), activation='relu',kernel_regularizer= regularizers.l2(0.001), input_shape = input_shape))
model.add(BatchNormalization())
model.add(Conv2D(64, kernel_size=(5,5), activation='relu',kernel_regularizer= regularizers.l2(0.001), input_shape = input_shape))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128,  kernel_regularizer= regularizers.l2(0.001),activation='relu', ))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))


model.compile(loss='categorical_crossentropy',
             optimizer= SGD(0.001, momentum=0.9),
             metrics = ['Accuracy'])

print(model.summary())
```

```python
batch_size = 32
epochs = 25

history = model.fit(x_train,
                    y_train,
                    batch_size = batch_size,
                    epochs = epochs,
                    verbose = 1,
                    validation_data = (x_test, y_test))

score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

### Adding augmentation then training

```python
train_datagen = ImageDataGenerator(
        rescale = 1./255,
        rotation_range=10,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        zoom_range=0.1,
        horizontal_flip=True,
        fill_mode='nearest')


batch_size = 32
epochs = 25

history = model.fit(train_datagen.flow(x_train, y_train, batch_size = batch_size),
                              epochs = epochs,
                              validation_data = (x_test, y_test),
                              verbose = 1,
                              steps_per_epoch = x_train.shape[0] // batch_size)

score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

**Training and Results:**
- The model is trained on the Fashion MNIST dataset using the Image Data Generator for data augmentation.
- Training accuracy and validation accuracy are monitored to assess model performance.
- The results show that regularization techniques help reduce overfitting and improve validation accuracy.
- Training with regularization may take longer, but it is expected to yield better results with more epochs.

```python
import matplotlib.pyplot as plt

history_dict = history.history

loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']

epochs = range(1, len(loss_values) + 1)

line1 = plt.plot(epochs, val_loss_values, label='Validation/Test Loss')
line2 = plt.plot(epochs, loss_values, label='Training Loss')
plt.setp(line1, linewidth=2.0, marker = '+', markersize=10.0)
plt.setp(line2, linewidth=2.0, marker = '4', markersize=10.0)
plt.xlabel('Epochs') 
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```
